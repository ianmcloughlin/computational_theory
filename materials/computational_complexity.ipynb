# Computational Complexity

## Time Complexity

> Let $M$ be a deterministic Turing machine that halts on all inputs.
> The running time or time complexity of $M$ is the function $f: \mathbb{N}_0 \rightarrow \mathbb{N}_0$, where $f(n)$ is the maximum number of steps that $M$ uses on any input of length $n$.
> If $f(n)$ is the running time of $M$, we say that $M$ runs in time $f(n)$ and that $M$ is an $f(n)$ time Turing machine.
> Customarily we use $n$ to represent the length of the input.  
> *- Sipser's [Introduction to the Theory of Computation](https://en.wikipedia.org/wiki/Introduction_to_the_Theory_of_Computation)*

## Functions

```python
# Range of n values.
n = np.arange(0, 21, 1)
```

```python
# A few examples of functions.
plt.plot(n, n, label='$n$')
plt.plot(n, 2*n, label='$2n$')
plt.plot(n, n+1, label='$n+1$')
plt.legend()
```

```python
# A few examples of functions.
plt.plot(n, n, label='$n$')
plt.plot(n, n**2, label='$n^2$')
plt.plot(n, 2**n, label='$2^n$')
plt.legend()
```

```python
m = np.arange(1, 201, 1)
plt.plot(m, m, label='$n$')
plt.plot(m, m**2, label='$n^2$')
plt.plot(m, m * np.log2(m), label='$n log n$')
plt.legend()
```

# Big-O Notation

> Let $f: \mathbb{N}_0 \rightarrow \mathbb{N}_0$ and $g: \mathbb{N}_0 \rightarrow \mathbb{N}_0$ be functions.
> We say that $f(n) \in O(g(n))$ if positive integers $c$ and $n^\prime$ exist such that for every integer $n \geq n^\prime$:
> $f(n) \leq cg(n).$  
> *- Edited from Sipser's [Introduction to the Theory of Computation](https://en.wikipedia.org/wiki/Introduction_to_the_Theory_of_Computation)*

## Example 1

$ f(n) = n $

$ g(n) = 2n $

$ f(n) \in O(g(n))$

$ n^\prime = 0 $

$ c = 1 $

$f(n) \leq cg(n) = 1 g(n) = g(n) = 2n$ for all $n > n^\prime = 0$

## Example 2

$ f(n) = 2n $

$ g(n) = n $

$ f(n) \in O(g(n))$

$ n^\prime = 0 $

$ c = 2 $

$f(n) = 2n \leq cg(n) = 2 g(n) = 2n$ for all $n > n^\prime = 0$

```python
n = np.arange(1, 21, 1)

plt.plot(n, n, label='$n$')
plt.plot(n, 2*n, label='$2n$')
plt.plot(n, 3*n, label='$cn = 3n$')
plt.legend()
```

## Conclusion

$ n \in O(2n) $ and $2n \in O(n)$

## Example 3

$ f(n) = 2n $

$ g(n) = n^2 $

Is $f(n) \in O(g(n))$?

Is $2n \in O(n^2)$?

$n^\prime = 2$

$c = 1$

$2n \leq cn^2 = 1n^2 = n^2$ for all $n > 2$

```python
n = np.arange(0, 3, 0.01)

plt.plot(n, 2*n, label='$2n$')
plt.plot(n, n**2, label='$n^2$')
plt.legend()
```

## Example 4

$f(n) = n^2$

$g(n) = 2^n$

Is $n^2 \in O(2^n)$?

$c = 1$

$n^\prime = 4$

$n^2 \leq 2^n$ for all $n > 4$

```python
n = np.arange(0, 5, 0.01)

plt.plot(n, n**2, label='$n^2$')
plt.plot(n, 2**n, label='$2^n$')
plt.legend()
```
